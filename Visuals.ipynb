```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df=spark.read.csv("/Volumes/workspace/default/retail_sales_dataset",header=True,inferSchema=True)
df=df.toPandas()
df.head(5)
```

```python
df.info()
```

```python
df["Date"]=pd.to_datetime(df["Date"],errors='coerce')
```

```python
df.info()
```

```python
df.columns
```

```python
revenue_by_product=df.groupby("Product Category")["Total Amount"].sum().reset_index(name="Revenue by Product Category")
revenue_by_product
```

```python
revenue_by_product=revenue_by_product.set_index("Product Category")
revenue_by_product.plot(kind="bar",title="Revenue by product Category",xlabel="Product Category",ylabel="Revenue by Product")
plt.xtricks(rotation = 45)
plt.show()
```
